```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read trading events
events = pd.read_csv('events.csv')
```

```python
# Plot AdjPnl and SP_Price
fig, ax1 = plt.subplots()
ax1.plot(events["ID"], events["AdjPnL"], color="tab:blue")

ax2 = ax1.twinx()
ax2.plot(events["ID"], events["SP_Price"], color="tab:red")
ax2.axhline(y=4050, color='g', linestyle='-')
ax2.axhline(y=4100, color='g', linestyle='-')
```
